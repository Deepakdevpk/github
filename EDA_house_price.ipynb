# ***IQR method***

```python
df2=df.copy()
```

```python
plt.subplot(2,2,1)
sns.boxplot(df2['price'])
plt.subplot(2,2,2)
sns.boxplot(df2['size'])
plt.subplot(2,2,3)
sns.boxplot(df2['total_sqft'])
plt.subplot(2,2,4)
sns.boxplot(df2['bath'])
```

```python
#define a function to calculate IQR
def detect_iqr_outliers(column):
  q1=column.quantile(0.25)
  q3=column.quantile(0.75)
  iqr=q3-q1
  lower_bound=q1-1.5*iqr
  upper_bound=q3+1.5*iqr
  return(column < lower_bound) | (column > upper_bound)

# Apply the function to numerical columns
outliers_iqr =df2[['size','total_sqft','bath','balcony','price']].apply(detect_iqr_outliers)

# view outliers
print('outliers detected using IQR:')
print(outliers_iqr.sum())
```

```python
from scipy.stats import zscore
z_score=df2[['size','total_sqft','bath','balcony','price']].apply(zscore)
threshold=3
outliers_z=(z_score.abs()> threshold)
print('outliers detected using z-score:')
print(outliers_z.sum())
```

# remove outliers

```python
# filter out the row with any outliers
df2=df2[~(outliers_iqr.any(axis=1))]
print('data after removing outliers (iqr):')
print(df2)
```

```python
plt.subplot(2,2,1)
sns.boxplot(df2['price'])
plt.subplot(2,2,2)
sns.boxplot(df2['size'])
plt.subplot(2,2,3)
sns.boxplot(df2['total_sqft'])
plt.subplot(2,2,4)
sns.boxplot(df2['bath'])
```

# ***Traning The Modeling***

```python
x_pca=df_pca.drop(columns=['price_log'],axis=1)
y=df_pca['price_log']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
# MAE(mean absolute error)
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
# MSE(Mean Squared Error)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
# RMSE(Root Mean Squared Error)
rmse=np.sqrt(mse)
rmse
```

```python
# R2 Score(R-squared)
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```
